```python
#Q4: Which are the top 10 countries with most gold, silver, bronze medal?


df.groupby('region')['Medal'].count().nlargest(10).reset_index()
```

```python
#Q5: Find the distribution of gold medalists according to their age.

goldMedals = df[(df.Medal == 'Gold')]
#print(goldMedals.head())

plt.figure(figsize=(20, 15))
plt.title('Distribution of Gold Medals')
sns.countplot(goldMedals['Age'])
plt.show()
```

```python
#Q8: Show a scatter plot for gold winners according to their age and weight. What do you see?


gold_winner= df[(df['Medal']=="Gold")]
gold_winner
```
